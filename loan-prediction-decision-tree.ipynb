## <div style="font-family: Trebuchet MS; background-color: #153B50; color: #FFFFFF; padding: 12px; line-height: 1.5;">Apa itu Pinjaman?</div>

Pinjaman adalah sejumlah uang yang dipinjam dari seseorang atau lembaga keuangan dengan perjanjian untuk membayarnya kembali dengan bunga dalam jangka waktu tertentu. Definisi pinjaman dapat dijelaskan sebagai:
- Harta benda, uang, atau barang material lainnya yang diberikan kepada pihak lain.
- Ditukar dengan pengembalian nilai pinjaman di masa depan beserta bunga dan biaya keuangan lainnya.
- Dapat berupa jumlah tertentu satu kali atau diambil sebagai jalur kredit terbuka hingga batas tertentu.
- Pinjaman hadir dalam berbagai bentuk seperti: Pribadi, Komersial, Agunan, dan Tanpa agunan

**Pengertian Pinjaman**

Pinjaman adalah hutang yang diambil oleh individu atau suatu entitas. Pihak lain dalam transaksi ini disebut pemberi pinjaman, biasanya pemerintah, lembaga keuangan, atau perusahaan. Mereka meminjamkan sejumlah uang yang dibutuhkan kepada peminjam. Sebagai gantinya, peminjam setuju untuk membayar sejumlah ketentuan tertentu, termasuk biaya keuangan, bunga, dll., dengan uang yang awalnya dipinjam.

**Bagaimana Proses Pinjaman Bekerja**

Saat anda membutuhkan uang, anda dapat mengajukan pinjaman dari perusahaan atau bank. Anda akan diminta untuk memberikan perincian khusus seperti mengapa anda membutuhkan pinjaman, anda harus mengungkapkan riwayat keuangan anda, nomor jaminan sosial, dan informasi lainnya yang dapat bervariasi dari pemberi pinjaman ke pemberi pinjaman.

Pemberi pinjaman akan meninjau aplikasi anda dan memeriksa rasio debt-to-come (DTI) untuk mengevaluasi apakah anda dapat membayar kembali pinjaman kepada mereka. Berdasarkan aplikasi anda, pemberi pinjaman akan menyetujui atau menolak aplikasi. Jika aplikasi anda ditolak, pemberi pinjaman harus memberikan alasan perihal penolakan.

Jika aplikasi anda disetujui, kontrak ditandatangani antara anda dan pemberi pinjaman. Pemberi pinjaman mentransfer jumlah pinjaman ke rekening anda yang perlu anda bayar kembali bersama dengan bunga dan biaya lainnya.

Ada persyaratan pinjaman yang harus disetujui oleh kedua belah pihak sebelum kontrak ditandatangani dan uang dicairkan. Dalam beberapa kasus, pemberi pinjaman memerlukan jaminan, yang detailnya tercantum dalam dokumen pinjaman. Sebagian besar pinjaman juga memiliki ketentuan mengenai jumlah bunga maksimum, dan jangka waktu sebelum pembayaran diperlukan.

**Mengapa dan Kapan Pinjaman Diberikan**

Pinjaman diberikan untuk berbagai alasan. Peminjam mungkin membutuhkan pinjaman untuk membeli barang, konsolidasi hutang, usaha bisnis, renovasi, atau investasi. Pinjaman bisnis dapat membantu perusahaan dalam memperluas operasi mereka.

Singkatnya, pinjaman memungkinkan pertumbuhan keseluruhan pasokan uang dalam perekonomian dan membuka persaingan dengan meminjamkan kepada bisnis baru. Bunga dan biaya lain yang diambil pemberi pinjaman (bank) dari peminjam merupakan sumber pendapatan bagi mereka.

**Faktor Penting Saat Mengambil Pinjaman**

Bagi peminjam, salah satu faktor terpenting saat mengambil pinjaman adalah suku bunga. Pinjaman dengan suku bunga lebih tinggi akan lebih mahal bagi peminjam- ia harus membayar cicilan bulanan yang lebih tinggi atau memakan waktu lebih lama untuk melunasi pinjaman dibandingkan dengan pinjaman dengan suku bunga yang lebih rendah. Misalnya, jika anda meminjam Rp.$5$ Jt dengan cicilan 5 tahun untuk pinjaman jangka waktu dengan suku bunga $4,5\%$, anda harus membayar cicilan bulanan sebesar Rp.$93.220$ selama 5 tahun ke depan. Namun, jika suku bunga anda adalah $9\%$, anda harus membayar Rp.$103.790$ untuk periode yang sama.

Mari kita pahami secara berbeda. Jika anda telah mengambil pinjaman sebesar Rp.$10$ Jt dan suku bunganya adalah $6\%$ dan anda memutuskan untuk membayar Rp.$200.000$ per bulan, anda perlu membayar selama *58* bulan untuk melunasi pinjaman atau saldo anda. Dengan bunga $20\%$, dan saldo serta pembayaran bulanan yang sama, dibutuhkan *108* bulan untuk melunasi pinjaman.

## <div style="font-family: Trebuchet MS; background-color: #153B50; color: #FFFFFF; padding: 12px; line-height: 1.5;">Perilaku Pelanggan</div>

Perilaku pelanggan mengacu pada kebiasaan membeli individu, termasuk tren sosial, pola frekuensi, dan faktor latar belakang yang mempengaruhi keputusan mereka untuk membeli sesuatu. Bisnis mempelajari perilaku pelanggan untuk memahami target audiens mereka dan menciptakan produk dan penawaran layanan yang lebih menarik.

Perilaku pelanggan tidak menggambarkan siapa yang berbelanja di toko anda, tetapi bagaimana mereka berbelanja di toko anda. Ini mengulas faktor-faktor seperti frekuensi belanja, preferensi produk, dan bagaimana penawaran pemasaran, penjualan, dan layanan anda diterima. Memahami detail ini membantu bisnis berkomunikasi dengan pelanggan dengan cara yang produktif dan menyenangkan.

Ada tiga faktor yang mempengaruhi perilaku pelanggan: pribadi, psikologis, dan sosial. Mari kita bahas masing-masing jenisnya.

## <div style="font-family: Trebuchet MS; background-color: #153B50; color: #FFFFFF; padding: 12px; line-height: 1.5;">Apa itu Analisis Perilaku Pelanggan?</div>

Analisis perilaku pelanggan adalah pengamatan kualitatif dan kuantitatif tentang bagaimana pelanggan berinteraksi dengan perusahaan anda. Anda mulai dengan mengelompokkan pelanggan ke dalam persona pembeli berdasarkan kesamaan minat mereka. Kemudian, amati setiap kelompok pada tahap masing-masing dalam perjalanan pelanggan untuk melihat bagaimana persona yang berbeda berinteraksi dengan perusahaan anda.

Analisis ini memberikan wawasan tentang variabel yang mempengaruhi audiens anda serta motif, prioritas, dan metode pengambilan keputusan yang dipertimbangkan pelanggan selama perjalanan mereka. Ini juga membantu anda memahami bagaimana perasaan pelanggan tentang perusahaan anda dan apakah persepsi itu sesuai dengan nilai-nilai inti mereka.

Dengan melakukan analisis perilaku pelanggan, anda dapat:
- Mengembangkan strategi pemasaran dan penjualan yang lebih efektif.
- Meningkatkan pengalaman pelanggan di seluruh perjalanan mereka.
- Membangun hubungan yang lebih kuat dengan pelanggan anda.
- Membuat keputusan bisnis yang lebih baik berdasarkan data dan wawasan.

## <div style="font-family: Trebuchet MS; background-color: #153B50; color: #FFFFFF; padding: 12px; line-height: 1.5;">Masalah Dataset dalam Prediksi Kredit Macet</div>

Sebuah organisasi ingin **memprediksi siapa kemungkinan gagal bayar** untuk produk pinjaman konsumen. Mereka memiliki data tentang perilaku historis pelanggan berdasarkan pengamatan mereka sebelumnya. Oleh karena itu, ketika mereka mendapatkan pelanggan baru, mereka ingin **memprediksi siapa yang lebih berisiko dan siapa yang tidak**.

Berikut ini adalah beberapa masalah potensial yang mungkin dihadapi organisasi dalam dataset mereka:
- **Data yang tidak lengkap**: Data yang hilang atau tidak lengkap dapat menyebabkan bias dan mengurangi akurasi model prediksi.
- **Data tidak seimbang**: Jika sebagian besar data berasal dari pelanggan yang membayar pinjaman tepat waktu, maka model mungkin kesulitan mengidentifikasi calon gagal bayar.
- **Data bias**: Data yang bias dapat menyebabkan model yang diskriminatif, yang dapat merugikan sebagian kelompok pelanggan.
- **Fitur yang tidak relevan**: Adanya fitur yang tidak relevan dalam dataset dapat menurunkan kinerja model prediksi.
- **Kebocoran data**: Kebocoran data terjadi ketika data yang digunakan untuk melatih model mengandung informasi tentang variabel target (misalnya, apakah pelanggan telah gagal bayar atau tidak). Kebocoran data dapat menyebabkan model yang terlalu optimis dan tidak akurat.

## <div style="font-family: Trebuchet MS; background-color: #153B50; color: #FFFFFF; padding: 12px; line-height: 1.5;">Cara Mendapatkan Pinjaman Pribadi</div>

Tidak ada formula pasti untuk memenuhi syarat mendapatkan pinjaman pribadi karena situasi keuangan setiap pemohon berbeda dan unik. Namun, ada beberapa aturan umum dan rekomendasi yang dapat membantu anda meningkatkan peluang untuk memenuhi syarat mendapatkan pinjaman pribadi.

Sebagian besar pemberi pinjaman pribadi meninjau nilai kredit, riwayat kredit, pendapatan, dan rasio utang terhadap pendapatan (DTI) anda untuk menentukan kelayakan anda. Meskipun persyaratan minimum untuk masing-masing faktor ini berbeda untuk setiap pemberi pinjaman, berikut beberapa rekomendasi:

- **Nilai kredit minimum 670** <br>
Mempertahankan nilai kredit minimal 670 akan meningkatkan peluang Anda untuk memenuhi syarat. Namun, jika anda ingin mendapatkan persyaratan paling menguntungkan, kami merekomendasikan nilai minimal 720.
- **Pendapatan bulanan yang konsisten dan stabil** <br>
Persyaratan pendapatan minimum dapat sangat bervariasi antar pemberi pinjaman, dengan beberapa bahkan tidak memiliki persyaratan. Namun, penting untuk memiliki pendapatan yang konsisten dan stabil minimal untuk menunjukkan bahwa anda mampu membayar cicilan bulanan anda.
- **Rasio DTI kurang dari $36\%$** <br>
Meskipun beberapa pemberi pinjaman akan menyetujui pemohon yang sangat memenuhi syarat dengan rasio hingga $50\%$, sebaiknya membuat target rasio DTI kurang dari $36\%$ untuk meningkatkan peluang anda memenuhi syarat.

Karena setiap pemberi pinjaman memiliki persyaratan minimum sendiri, sebaiknya anda melakukan pra-kualifikasi jika mungkin dan konfirmasi dengan pemberi pinjaman tentang tolak ukur apa yang perlu anda penuhi. Ini akan memastikan anda hanya mengajukan pinjaman yang sesuai dengan situasi keuangan anda yang spesifik.

```python
def numerical_plot(data, col, title, simb, ylabel, color):
    fig, ax = plt.subplots(
        2, 1,
        sharex= True,
        figsize= (20, 8),
        facecolor= "#FFFFFF",
        gridspec_kw= {"height_ratios": (.2, .8)}
    )
    
    ax[0].set_facecolor("#FFFFFF")
    ax[1].set_facecolor("#FFFFFF")
    
    ax[0].set_title(
        title,
        weight= "bold",
        fontsize= 30,
        pad= 30,
    )
    
    sns.boxplot(
        x= col,
        data= data,
        color= color,
        ax= ax[0],
    )
    
    ax[0].set(yticks= [])
    
    sns.distplot(data[col], kde= True, color= color)
    
    plt.xticks(weight= "bold", fontsize= 10)
    plt.yticks(weight= "bold", fontsize= 10)
    
    ax[0].set_xlabel(col, weight= "bold", fontsize= 15, labelpad= 15)
    ax[1].set_xlabel(col, weight= "bold", fontsize= 15)
    ax[1].set_ylabel(ylabel, weight= "bold", fontsize= 15)
    
    plt.axvline(
        data[col].mean(),
        color= "darkblue",
        linewidth= 2.2,
        label= "mean=" + str(np.round(data[col].mean(), 1)) + simb,
    )
    
    plt.axvline(
        data[col].median(),
        color= "red",
        linewidth= 2.2,
        label= "median=" + str(np.round(data[col].median(), 1)) + simb,
    )
    
    plt.axvline(
        data[col].max(),
        color= "blue",
        linewidth= 2.2, 
        label= "max=" + str(np.round(data[col].max(), 1)) + simb,
    )
    
    plt.axvline(
        data[col].min(), 
        color= "orange", 
        linewidth= 2.2, 
        label= "min=" + str(np.round(data[col].min(), 1)) + simb,
    )
    
    plt.axvline(
        data[col].mode()[0], 
        color= "purple", 
        linewidth= 2.2, 
        label= "mode=" + str(data[col].mode()[0]) + simb,
    )
    
    plt.legend(
        bbox_to_anchor= (1, 1),
        ncol= 1,
        fontsize= 17,
        fancybox= True,
        shadow= True,
        frameon= False,
    )
    
    plt.tight_layout()
    plt.show()
```

```python
def bivariate_numerical_plot(data, x, hue, title, xlabel, ylabel):
    fig, ax = plt.subplots(
        1, 1,
        figsize= (20, 6),
        facecolor= "#FFFFFF"
    )
    
    sns.kdeplot(
        data= data, x= x,
        hue= hue, fill= True,
    )
    
    ax.set_facecolor("#FFFFFF")
    
    plt.title(title, weight= "bold", fontsize= 25, pad= 30)
    plt.xticks(weight= "bold", fontsize= 10)
    plt.yticks(weight= "bold", fontsize= 10)
    plt.xlabel(xlabel, weight= "bold", fontsize= 15, labelpad= 15)
    plt.ylabel(ylabel, weight= "bold", fontsize= 15, labelpad= 15)
    plt.tight_layout()
    plt.show()
```

## <div style="font-family: Trebuchet MS; background-color: #153B50; color: #FFFFFF; padding: 12px; line-height: 1.5;">Stats Function</div>

```python
def spearman_corr(data1, data2, title):
    print(f"-- {title} --")
    print()
    
    coef, p = stats.spearmanr(data1, data2)
    print(f"Spearmans corr coef: {coef:.3f}")
    print()
    
    alpha = .05
    if (p > alpha):
        print(f"Sampel tidak berkorelasi (gagal untuk menolak H0) p= {p:.3f}")
    else:
        print(f"Sampel memiliki korelasi (ditemukan bukti menolak H0) p= {p:.3f}")
    print()
```

```python
def pearson_corr(data1, data2, title):
    print(f"-- {title} --")
    print()
    
    coef, p = stats.pearsonr(data1, data2)
    print(f"Pearson corr coef: {coef:.3f}")
    print()
    
    alpha = .05
    if (p > alpha):
        print(f"Sampel tidak berkorelasi (gagal untuk menolak H0) p= {p:.3f}")
    else:
        print(f"Sampel memiliki korelasi (ditemukan bukti menolak H0) p= {p:.3f}")
    print()
```

```python
def skew_kurtosis(data, title):
    skw = stats.skew(data, bias= False)
    kurt = stats.kurtosis(data, bias= False)
    
    print(f"-- {title} --")
    print()
    
    print(f"Skewness: {skw}")
    if (skw < -1) and (skw > 1):
        print("Sangat Condong")
    elif (-1 < skw < -.5) or (1 < skw < .5):
        print("Cukup Condong")
    elif (-.5 < skw < .5):
        print("Kurang Lebih Simetris")
    print("-" * 35)
    
    print(f"Kurtosis: {kurt}")
    if (kurt > 1):
        print("Distribusinya terlalu memuncak")
    elif (kurt < -1):
        print("Distribusinya terlalu datar")
    else:
        print("Distribusinya terlalu tidak normal")
    print("-" * 35)
```

```python
def chebychev(data, title, interpret):
    mu = round(data.mean(), 1)
    sd = round(data.std(), 1)
    
    fc = round(mu - (2 * sd), 1)
    sc = round(mu + (2 * sd), 1)
    
    print(f"-- Interpretasi dari {title} --")
    print(f"Perhitungan pertama: {fc}")
    print(f"Perhitungan kedua: {sc}")
    print(interpret)
    print("-" * 35)
```

## <div style="font-family: Trebuchet MS; background-color: #153B50; color: #FFFFFF; padding: 12px; line-height: 1.5;">Analisis Univariate</div>

<p style="font-name: monospace; line-height: 2; font-size: 20px; font-weight: bold; letter-spacing: 2px; text-align: center;">Teorema Chebyshev</p>

$$
\begin{aligned}
1 - \frac{1}{k^2}: k &= 2 -> 1 - \frac{1}{2^2} = \frac{3}{4} -> 75 \\
                   k &= 3 -> 1 - \frac{1}{3^2} = \frac{8}{9} -> 88.9
\end{aligned}
$$

<br>

<p style="font-name: monospace; line-height: 2; font-size: 20px; font-weight: bold; letter-spacing: 2px; text-align: center;">Cara Menghitung Standar Deviasi ($\sigma$)?</p>
<p style="font-name: monospace; line-height: 2; font-size: 13px; font-weight: bold; letter-spacing: 2px; text-align: center;">Berikut Persamaannya:</p>

$$
\begin{aligned}
\sigma &= \sqrt{\sigma^2} = \sqrt{\frac{\sum{(x - \mu)^2}}{N}} \\
s &= \sqrt{s^2} = \sqrt{\frac{\sum{(x - \bar{x})^2}}{n - 1}}
\end{aligned}
$$

<br>

<p style="font-name: monospace; line-height: 2; font-size: 20px; font-weight: bold; letter-spacing: 2px; text-align: center;">Menghitung Mean ($\mu$)?</p>
<p style="font-name: monospace; line-height: 2; font-size: 13px; font-weight: bold; letter-spacing: 2px; text-align: center;">Berikut Persamaannya:</p>

$$
\begin{aligned}
\mu = \frac{\sum{x}}{N} \\
\bar{x} = \frac{\sum{x}}{n}
\end{aligned}
$$

<br>

<p style="font-name: monospace; line-height: 2; font-size: 20px; font-weight: bold; letter-spacing: 2px; text-align: center;">Notasi</p>
<p style="font-name: monospace; line-height: 2; font-size: 13px; font-weight: bold; letter-spacing: 2px; text-align: center;">Review Notasi:</p>

| Populasi | Sampel | Keterangan |
| --- | --- | --- |
| $\sigma^2$ | $s^2$ | Varians |
| $\sigma$ | $s$ | Standar Deviasi |
| $\mu$ | $\bar{x}$ | Mean |
| $N$ | $n$ | Sum of Entry (SOE) |
| $x - \mu$ | $x - \bar{x}$ | Deviasi |
| $\sum{(x - \mu)^2}$ | $\sum{(x - \bar{x})^2}$ | Sum of Square (SS) |

<br>

<p style="font-name: monospace; line-height: 2; font-size: 20px; font-weight: bold; letter-spacing: 2px; text-align: center;">Tentang skewness:</p>

<p style="font-name: monospace; line-height: 2; font-size: 13px; font-weight: normal; letter-spacing: 2px; text-align: center;">Sebagai pedoman umum, skewness dapat diartikan sebagai berikut: Jika skewness kurang dari $-1$ atau lebih dari $1$, distribusi data sangat miring. Ini berarti distribusi data menyimpang jauh dari bentuk simetris. Jika skewness antara $-1$ dan $-0.5$ atau antara $0.5$ dan $1$, distribusi data miring sedang. Ini berarti distribusi data masih menunjukkan ketidaksimetrisan, tetapi tidak terlalu ekstrim. Jika skewness antara $-0.5$ dan $0.5$, distribusi data kira-kira simetris. Ini berarti distribusi data mendekati bentuk normal, dengan sisi kiri dan kanan kurva berbentuk cermin.</p>

<br>

<p style="font-name: monospace; line-height: 2; font-size: 20px; font-weight: bold; letter-spacing: 2px; text-align: center;">Tentang kurtosis:</p>

<p style="font-name: monospace; line-height: 2; font-size: 13px; font-weight: normal; letter-spacing: 2px; text-align: center;">Untuk kurtosis, pedoman umum adalah sebagai berikut: Jika nilai kurtosis lebih besar dari $+1$, distribusi data terlalu puncak (puncaknya terlalu tinggi). Ini berarti distribusi memiliki data yang lebih terpusat di sekitar nilai tengah dibandingkan dengan distribusi normal. Jika nilai kurtosis kurang dari $-1$, distribusi data terlalu datar (puncaknya terlalu rendah). Ini berarti data lebih menyebar dibandingkan dengan distribusi normal.</p>

## <div style="font-family: Trebuchet MS; background-color: #153B50; color: #FFFFFF; padding: 12px; line-height: 1.5;">Analisis Bivariat</div>

### <div style="font-family: Trebuchet MS; background-color: #153B50; color: #FFFFFF; padding: 12px; line-height: 1.5;">Apa itu Korelasi?</div>

**Korelasi: Hubungan Antar Variabel dalam Data**

Korelasi adalah ukuran statistik (diungkapkan dalam angka) yang menggambarkan keakuratan dan arah hubungan antara dua atau lebih variabel. Namun, adanya korelasi antara variabel tidak serta merta berarti perubahan pada satu variabel merupakan penyebab perubahan nilai variabel lainnya.

Variabel dalam kumpulan data dapat berhubungan karena berbagai alasan.

Misalnya:
- Satu variabel dapat menyebabkan atau tergantung pada nilai variabel lain.
- Satu variabel dapat sedikit berhubungan dengan variabel lain.
- Dua variabel dapat bergantung pada variabel ketiga yang tidak diketahui.

Dalam analisis data dan pemodelan, memahami hubungan antar variabel sangatlah penting. Hubungan statistik antara dua variabel disebut sebagai korelasi mereka.

Korelasi bisa positif, artinya kedua variabel bergerak ke arah yang sama, atau negatif, artinya ketika nilai satu variabel meningkat, nilai variabel lain menurun. Korelasi juga bisa netral atau nol, artinya kedua variabel tidak berhubungan.

- **Korelasi Positif:** kedua variabel berubah ke arah yang sama.
- **Korelasi Netral:** tidak ada hubungan dalam perubahan variabel.
- **Korelasi Negatif:** variabel berubah ke arah yang berlawanan.

Performa beberapa algoritma dapat menurun jika dua atau labih variabel terkait erat, yang disebut multikolinearitas. Contohnya adalah regresi linier, di mana salah satu variabel berkorelasi yang mengganggu harus dihapus untuk meningkatkan keterampilan model.

Kita mungkin juga tertarik dengan korelasi antara variabel input dengan variabel output untuk memberikan gambaran tentang variabel mana yang mungkin atau tidak relevan sebagai input untuk mengembangkan model.

Struktur hubungan mungkin diketahui, misalnya bersifat linier, atau kita mungkin sama sekali tidak tahu apakah suatu hubungan antara dua variabel ada atau seperti apa strukturnya. Bergantung pada apa yang diketahui tentang hubungan dan distribusi variabel, skor korelasi yang berbeda dapat dihitung.

**Korelasi Pearson**

$$
r = \frac{\sum(x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum(x_i - \bar{x})^2 \sum(y_i - \bar{y})^2}}
$$

**Koefisien Korelasi Pearson: Mengukur Kekuatan Hubungan Linear Antar Data**

Koefisien korelasi Pearson (diambil dari nama Karl Pearson) dapat digunakan untuk meringkas kekuatan hubungan linear antara dua sampel data.

Nilai koefisien ini dihitung sebagia kovarians dari dua variabel dibagi dengan produk deviasi standar dari masing-masing sampel data. Ini merupakan normalisasi kovarians antara dua variabel untuk memberikan skor yang dapat diinterpretasikan.

**Apa maksudnya?**
- Kovarians: Mengukur bagaimana dua variabel cenderung bergerak "bersama-sama".
- Deviasi standar: Mengukur seberapa tersebar data di sekitar nilai rata-ratanya.
- Normalisasi: Mengubah nilai menjadi kisaran tertentu (biasanya -1 hingga 1) untuk memudahkan perbandingan.

Dengan kata lain, koefisien korelasi Pearson memberi Anda angka antara -1 dan 1 yang menunjukkan kekuatan dan arah hubungan linear antara dua variabel.

- **-1: Korelasi negatif sempurna.** Kedua variabel bergerak berlawanan arah.
- **0: Tidak ada korelasi linear.** Kedua variabel tidak terkait.
- **+1: Korelasi positif sempurna.** Kedua variabel bergerak searah.

Nilai yang lebih dekat ke -1 atau +1 menunjukkan hubungan yang lebih kuat, sedangkan nilai yang lebih dekat ke 0 menunjukkan hubungan yang lebih lemah.

> `Koefisien korelasi Pearson = covariance(X, Y) / (stdv(X) * stdv(Y))`

Penggunaan mean dan deviasi standar dalam perhitungan koefisien korelasi Pearson menunjukkan bahwa kedua sampel data sebaiknya memiliki distribusi normal (Gaussian) atau mirip Gaussian.

**Interpretasi hasil:**

Hasil perhitungan, yaitu koefisien korelasi, dapat digunakan untuk memahami hubungan antara dua variabel. Koefisien ini memiliki nilai antara $-1$ dan $1$, yang mewakili batas korelasi dari korelasi negatif penuh hingga korelasi positif penuh. Nilai $0$ berarti tidak ada korelasi.

Interpretasi nilai koefisien korelasi:
- Nilai di bawah $-0.5$ atau di atas $0.5$ sering menunjukkan korelasi yang nyata.
- Nilai di bawah kisaran tersebut menunjukkan korelasi yang kurang nyata.

Fungsi **pearsonr()** dalam SciPy dapat digunakan untuk menghitung koefisien korelasi Pearson antara dua sampel data dengan panjang yang sama.

### <div style="font-family: Trebuchet MS; background-color: #153B50; color: #FFFFFF; padding: 12px; line-height: 1.5;">Korelasi Rank Nonparametrik</div>

Korelasi mengacu pada hubungan antara nilai yang diamati dari dua variabel. Hubungan ini bisa:

- **Positif:** Ketika nilai satu variabel meningkat, nilai variabel lain juga meningkat.
- **Negatif:** Ketika nilai satu variabel meningkat, nilai variabel lain menurun.
- **Netral:** Tidak ada hubungan antara kedua variabel.

Korelasi mengukur hubungan ini, biasanya dengan nilai antara $-1$ dan $1$. Nilai $-1$ menunjukkan korelasi negatif sempurna, dan $1$ menunjukkan korelasi positif sempurna. Nilai $0$ menunjukkan tidak ada korelasi.

**Tabel: Kekuatan Hubungan Antar Variabel**
| Koefisien Korelasi Hubungan Langsung | Koefisien Korelasi Hubungan Tidak Langsung | Kekuatan Hubungan Antar Variabel |
| 0.0 | 0.0 | Tidak ada/trivial |
| 0.1 | -0.1 | Lemah/kecil |
| 0.3 | -0.3 | Moderat/sedang |
| 0.5 | -0.5 | Kuat/besar |
| 1.0 | -1.0 | Sempurna |

Korelasi antar variabel dapat dihitung menggunakan metode standar seperti korelasi Pearson, namun hanya jika datanya berdistribusi normal (Gaussian). Untuk data yang tidak normal, diperlukan **korelasi rank nonparametrik**.

Korelasi rank mengukur hubungan antar variabel berdasarkan urutan data, bukan nilai spesifiknya. Data ordinal memiliki label dan urutan, misalnya **‘rendah‘**, **‘sedang‘**, dan **‘tinggi‘**.

Korelasi rank dapat dihitung untuk data numerik dengan mengubah nilainya menjadi urutan (rank). Koefisien korelasi rank kemudian dihitung untuk mengukur hubungan antar rank.

Karena tidak bergantung pada asumsi distribusi, metode ini disebut korelasi bebas distribusi atau nonparametrik. Korelasi rank juga sering digunakan sebagai dasar uji hipotesis statistik lainnya, seperti menentukan apakah dua sampel berasal dari populasi yang sama atau berbeda.

Beberapa contoh metode korelasi rank:
- Korelasi Spearman
- Korelasi Kendall
- Korelasi Goodman dan Kruskal
- Korelasi Somers

**Korelasi Rank Spearman**

$$
p = 1 - \frac{6 \sum d^2_i}{n(n^2 - 1)}
$$

Korelasi rank Spearman dinamai Charles Spearman dan dilambangkan dengan huruf Yunani kecil rho (ρ).

Metode ini mengukur sejauh mana dua variabel peringkat berasosiasi dengan fungsi monotonik (meningkat atau menurun). Sebagai uji hipotesis, metode ini berasumsi bahwa sampel tidak berkorelasi (gagal menolak H0).

Intuisinya adalah bahwa korelasi rank Spearman menghitung korelasi Pearson menggunakan nilai rank, bukan nilai aktual. Korelasi Pearson adalah perhitungan kovariansi antara dua variabel, dinormalisasi oleh variansnya.

Korelasi rank Spearman dapat dihitung dalam Python menggunakan fungsi **spearmanr()** dari SciPy.

```python
# Definisikan kolom dengan tipe obyek untuk menerapkan LabelEncoder
col_object = loan.select_dtypes("object").columns.tolist()

encoder = LabelEncoder() # Object LabelEncoder
for col in col_object:
    loan[col] = encoder.fit_transform(loan[col].values.tolist())
```

```python
# Mutual information tentang data
X_mutual = loan.astype("int64").reset_index().copy()
y_mutual = X_mutual.pop("risk_flag")

del X_mutual["Id"]

# Diskritisasi fitur sehingga sekarang menjadi tipe data integer
discret_features = X_mutual.dtypes == int

# Mutual Information Scores Function
def calc_mi(X_mutual, y_mutual, discret_features):
    mi_scores = mutual_info_classif(X_mutual, y_mutual, discrete_features= discret_features)
    mi_scores = pd.Series(mi_scores, name= "MI Scores", index= X_mutual.columns)
    mi_scores = mi_scores.sort_values(ascending= False)
    return mi_scores

mi_scores = calc_mi(X_mutual, y_mutual, discret_features)
mi_scores.to_frame()
```

```python
# Variabel
scores = mi_scores.sort_values(ascending= True)
width = np.arange(len(scores))
ticks = list(scores.index)

# Plotting
plt.figure(figsize= (15, 5))
plt.barh(width, scores, color= "gold", alpha= .7, edgecolor= "k")
plt.yticks(width, ticks, weight= "bold")
plt.xticks(weight= "bold")
plt.title("Score Mutual Information", weight= "bold", fontsize= 20, pad= 25)
```

```python
# Function untuk plotting confusion matrix
def plot_cm(
    cm, classes, normalize= False,
    title= "Confusion Matrix", cmap= plt.cm.Blues
):
    # Buat plot figure-nya
    plt.figure(figsize= (10, 10))
    plt.imshow(cm, interpolation= 'nearest', cmap= cmap)
    plt.title(title)
    plt.colorbar()

    # Beri label pada garis x dan y
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation= 45)
    plt.yticks(tick_marks, classes)

    # Cek kondisi normalisasi
    if normalize:
        cm = cm.astype("float") / cm.sum(axis= 1)[:, np.newaxis]
        print("Normalized Confusion Matrix")
    else:
        print("Confusion Matrix, without Normalization")

    # Threshold
    thresh = cm.max() / 2.
    # Bentuk plot confusion matrix-nya
    for i, j in itertools.product(
        range(cm.shape[0]), 
        range(cm.shape[1])
    ):
        plt.text(
            j, 
            i, 
            cm[i, j], 
            fontsize= 16, 
            horizontalalignment= "center", 
            color= "white" if cm[i, j] > thresh else "black"
        )

    plt.tight_layout()
    plt.ylabel("True Labels")
    plt.xlabel("Predicted Labels")
    plt.show()
    
cm = confusion_matrix(y_test, y_pred)
classes = np.unique(y_test)
plot_cm(cm, classes)
```
